```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = r'http://en.wikipedia.org/wiki/Anna_Karina'
tabele = pd.read_html(url)
df = tabele[1]
df.columns = ['data', 'titlul', 'rolul', 'regizorul']
df = df.drop(df.head(1).index)
```

```python
df.fillna(axis = 0, method = 'ffill', inplace  = True)
```

```python
def conversie(y):
    return y.year
df['data'] = df['data'].apply(conversie)
```

```python
uniques = df['regizorul'].unique()
```

```python

df.head()
```

```python
for value in uniques:
    df[value] = (df['regizorul'] == value) 
    """
    Adauga cate o coloana pt fiecare regizor cu 'true' daca regirul respectiv a regizat filmul de pe fiecare rand si false in rest
    """
```

```python
grouped = df.ix[:,'Jean-Luc Godard':'Anna Karina herself'].groupby(df['data']).sum() #nr de filme din fiecare an, pt fiecare regizor 
```

```python
grouped.head()
```

```python
grouped.sum() >= 4
```

```python
reduced = grouped[['Jean-Luc Godard', 'Roger Vadim', 'Jacques Rivette', 'Luchino Visconti']]
```
